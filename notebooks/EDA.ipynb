```python
import pandas as pd

df = pd.read_csv('../data/diabetes.csv')
df.head()
```

## Diabetic vs Non-diabetic

```python
import seaborn as sns
import matplotlib.pyplot as plt

custom_palette = {'0': 'skyblue', '1': 'salmon'}

sns.countplot(x='Outcome', data=df, palette=custom_palette)
plt.title("Diabetes Outcome Distribution")
plt.xticks([0, 1], ['No Diabetes', 'Diabetic'])
plt.savefig("../output/class_distribution.png")
plt.show()
```

## Corelation between different features towards prediction

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Features")
plt.savefig("../output/correlation_heatmap.png")
plt.show()
```

## Distribution of BMI in the population

```python
sns.histplot(df['BMI'], kde=True)
plt.title("BMI Distribution")
plt.savefig("../output/bmi_dist.png")
```
